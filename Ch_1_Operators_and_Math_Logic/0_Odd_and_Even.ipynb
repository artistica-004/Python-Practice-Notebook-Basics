```python
n=int(input("Enter the number"))
ndigit=len(str(n))
original=n
total=0
while n>0:
        ldigit=n%10
        total+=ldigit**ndigit
        n=n//10
print(total)
if total==original:
        print("Number is Armstrong")
else:
        print("Number is not Armstrong")
```
