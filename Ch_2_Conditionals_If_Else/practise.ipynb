```python
h = float(input("Enter the height: "))
w = float(input("Enter the weight: "))

bmi = w / (h ** 2)

if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"BMI: {round(bmi, 2)} ({category})")
```

```python
unit=float(input("Enter the unit"))
bill=0
if unit<=100:
    bill=unit*1.5
elif unit<=200:
    bill=100*1.5+(unit-100)*2.5
elif unit<=300:
    bill=100*1.5+100*2.5+(unit-200)*4
else:
    bill=100*1.5+100*2.5+100*4+(unit-300)*6
    
print(f"Electricity Bill: ₹{bill}")

    
```

```python
n1=int(input("Enter n1"))
n2=int(input("Enter n2"))
operator = input("Enter operator (+,-,*,/,%): ")
sum=n1+n2
diff=n1-n2
product=n1*n2
div=n1/n2
mod=n1%n2
if operator=="+":
    print(sum)
elif operator=="-":
    print(diff)
elif operator=="*":
    print(product)
elif operator=="/":
    print(div)    
elif operator=="%":
    print(mod)
else:
    print("Invalid operator")
```
